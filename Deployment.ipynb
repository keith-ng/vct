```python
# Function to locate frame based on inputs
def find_frame(map_num, round_num, time, spike_planted):
    
    full = pd.read_csv('liq_lev_m' + str(map_num) + '_md_final.csv')
    positions = pd.read_csv('liq_lev_m' + str(map_num) + '_pl_final.csv')
    
    if map_num == 1:
        game_map = 'assets/haven.png'
        agents_dict = {'p1': 'omen', 'p2':'fade', 'p3':'raze', 'p4':'breach', 'p5':'chamber',
                    'p6': 'omen', 'p7':'fade', 'p8':'raze', 'p9':'breach', 'p10':'chamber'}
        col_lst = ['p1a_cast', 'p1b_cast', 'p1c_cast', 'p2a_cast', 'p2b_cast', 
            'p2c_cast', 'p3a_cast', 'p3b_cast', 'p3c_cast', 'p4a_cast', 
            'p4b_cast', 'p4c_cast', 
            'p6a_cast', 'p6b_cast', 'p6c_cast', 'p7a_cast', 'p7b_cast',
            'p7c_cast', 'p8a_cast', 'p8b_cast', 'p8c_cast', 'p9a_cast', 
            'p9b_cast', 'p9c_cast']
        
    elif map_num == 2:
        game_map = 'assets/ascent.png'
        agents_dict = {'p1': 'omen', 'p2':'fade', 'p3':'phoenix', 'p4':'kayo', 'p5':'chamber',
                              'p6': 'astra', 'p7':'fade', 'p8':'kayo', 'p9':'sova', 'p10':'chamber'}
        col_lst = ['p1a_cast', 'p1b_cast', 'p1c_cast', 'p2a_cast', 'p2b_cast', 'p2c_cast', 
                   'p3a_cast', 'p3b_cast', 'p3c_cast', 'p4a_cast', 'p4b_cast', 'p4c_cast', 
                   'p7a_cast', 'p7b_cast', 'p7c_cast', 'p8a_cast', 'p8b_cast', 'p8c_cast', 
                   'p9a_cast', 'p9b_cast', 'p9c_cast']
    
    df = full
    if spike_planted == False:
        df = df[df['secs'].notna()]
        frame = df.loc[(df['round'] == round_num) & (df['secs'] == time), 'frame']
    else:
        df = df[df['secs_spike'].notna()]
        frame = df.loc[(df['round'] == round_num) & (df['secs_spike'] == time), 'frame'] 
    return frame
```

```python
# Function to plot map with icons
def minimap(map_num, round_num, time, spike_planted):
    
    full = pd.read_csv('liq_lev_m' + str(map_num) + '_md_final.csv')
    positions = pd.read_csv('liq_lev_m' + str(map_num) + '_pl_final.csv')
    
    if map_num == 1:
        game_map = 'assets/haven.png'
        agents_dict = {'p1': 'omen', 'p2':'fade', 'p3':'raze', 'p4':'breach', 'p5':'chamber',
                    'p6': 'omen', 'p7':'fade', 'p8':'raze', 'p9':'breach', 'p10':'chamber'}
        col_lst = ['p1a_cast', 'p1b_cast', 'p1c_cast', 'p2a_cast', 'p2b_cast', 
            'p2c_cast', 'p3a_cast', 'p3b_cast', 'p3c_cast', 'p4a_cast', 
            'p4b_cast', 'p4c_cast', 
            'p6a_cast', 'p6b_cast', 'p6c_cast', 'p7a_cast', 'p7b_cast',
            'p7c_cast', 'p8a_cast', 'p8b_cast', 'p8c_cast', 'p9a_cast', 
            'p9b_cast', 'p9c_cast']
        
    elif map_num == 2:
        game_map = 'assets/ascent.png'
        agents_dict = {'p1': 'omen', 'p2':'fade', 'p3':'phoenix', 'p4':'kayo', 'p5':'chamber',
                              'p6': 'astra', 'p7':'fade', 'p8':'kayo', 'p9':'sova', 'p10':'chamber'}
        col_lst = ['p1a_cast', 'p1b_cast', 'p1c_cast', 'p2a_cast', 'p2b_cast', 'p2c_cast', 
                   'p3a_cast', 'p3b_cast', 'p3c_cast', 'p4a_cast', 'p4b_cast', 'p4c_cast', 
                   'p7a_cast', 'p7b_cast', 'p7c_cast', 'p8a_cast', 'p8b_cast', 'p8c_cast', 
                   'p9a_cast', 'p9b_cast', 'p9c_cast']
        
    frame = find_frame(map_num, round_num, time, spike_planted)
    frame_num = frame.iloc[0]

    # iloc[0] for first frame of the second, change if necessary
    df = positions.loc[positions['frame'] == frame_num]
    df.reset_index(drop=True, inplace=True)
    
    # backup in case obj missed
    if full.loc[full['frame'] == frame_num]['round'].iloc[0] == full.loc[full['frame'] == frame_num +1]['round'].iloc[0]:
        df1 = positions.loc[positions['frame'] == frame_num +1]
        df1.reset_index(drop=True, inplace=True)

    img = Image.open(game_map)
    # specify icon size for convenience
    icon_size = (30,30)

    # Plot all detected objects
    for i in range(len(df)):
        x = df.loc[i]['x_fromleft'] * img.size[0] - (icon_size[0]/2)
        y = df.loc[i]['y_fromtop'] * img.size[1] - (icon_size[1]/2)

        icon = Image.open('assets/agents/' + df.loc[i]['obj_class'] + '.png')
        icon = icon.resize((icon_size[0],icon_size[1]),Image.ANTIALIAS)
        img.paste(icon, (round(x),round(y)), mask = icon)
    
    det_lst = df['obj_class'].unique().tolist()
    
    # Take object from frame + 1 if not detected in frame based on ground truth list
    # has to be +1 because if -1 the guy could've died but still appear
    # Repeat for +2 doesn't seem viable since 0.5secs can be significant & code will take longer
    for i in range(len(df1)):
        if df1.loc[i]['obj_class'] not in det_lst:
            try:
                x = df1.loc[i]['x_fromleft'] * img.size[0] - (icon_size[0]/2)
                y = df1.loc[i]['y_fromtop'] * img.size[1] - (icon_size[1]/2)

                icon = Image.open('assets/agents/' + df1.loc[i]['obj_class'] + '.png')
                icon = icon.resize((icon_size[0],icon_size[1]),Image.ANTIALIAS)
                img.paste(icon, (round(x),round(y)), mask = icon)
            except:
                pass
            
    img.save('mm_m'+str(map_num)+'_r'+ str(round_num)+'_t'+str(time)+'_sp'+str(spike_planted)+'.png')
#     return img
#     img.show()
```

```python
def locate_round(team, side, agent, rnd):
    frames_lst =[]
    pos_lst = []
    for i in range(len(full)):
        if full.loc[i]['round'] == rnd:
            frames_lst.append(full.loc[i]['frame'])
    for i in range(len(positions)):
        if positions.loc[i]['frame'] in frames_lst:
            if positions.loc[i]['team'] == team:
                if positions.loc[i]['side'] == side:
                    if positions.loc[i]['agent'] == agent:
                        pos_lst.append(positions.loc[i])
    df = pd.DataFrame(pos_lst)
    return df
```
